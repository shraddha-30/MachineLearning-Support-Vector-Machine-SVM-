```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv('Ecommerce Customers - Ecommerce Customers.csv')
customers.head()
```

```python
customers.info()
```

```python
customers.describe()
```

```python
import seaborn as sns
```

```python
sns.jointplot(customers['Time on Website'],customers['Yearly Amount Spent'])
plt.title('shraddha')
```

```python
sns.jointplot(customers['Time on App'],customers['Yearly Amount Spent'])
plt.title('shraddha')
```

```python
sns.pairplot(customers)
plt.title('shraddha')
```

```python
y= customers['Yearly Amount Spent']
x=customers[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
```

```python
sns.lmplot(x ='Yearly Amount Spent',y ='Length of Membership' ,data=customers)
plt.title('shraddha')
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(x_train, y_train)
print('Co-efficient:\n',lm.coef_)
```

```python
prediction=lm.predict(x_test)
```

```python
#Evaluating the model 
```

```python
plt.scatter(y_test, prediction)
plt.xlabel('y_test')
plt.ylabel('predicted y')
plt.title('shraddha')
```

```python
from sklearn import metrics
```

```python
#residuals
```

```python
print('MAE:',metrics.mean_absolute_error(y_test,prediction))

```

```python
print('MSE:',metrics.mean_squared_error(y_test,prediction))
```

```python
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```

```python
sns.displot((y_test,prediction),bins=50)
plt.title('shraddha')
```
